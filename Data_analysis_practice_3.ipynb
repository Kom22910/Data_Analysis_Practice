# Q14 Do you notice any outliers in the 'Age' column? How might you deal with outliers?

```python
plt.figure(figsize = (6,10))
df["age"].plot(kind = "box")
plt.show()
```

**or**

```python
plt.figure(figsize = (6,10))
sns.boxplot(df["age"])
plt.show()
```

**or**

```python
plt.boxplot(df["age"])
plt.show()
```

**or by IOR method**

```python
Q1 = df["age"].quantile(0.25)
Q3 = df["age"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outlier = df[(df["age"] < lower_bound) | (df["age"] > upper_bound)]
outlier.shape
```

**removable of outlier**

```python
no_outlier = df[(df["age"] > lower_bound) & (df["age"] <= upper_bound)]
no_outlier.shape
```

**or reduce by taking square root or log**

```python
df["age"] = np.log(df["age"])

plt.boxplot(df["age"])
plt.show()
```

# Q19 Visualize the relationship between education and income using a bar chart.

```python
df.columns
```

```python
education_income = df.groupby("education")["salary"].value_counts(normalize = True).unstack().fillna(0)
education_income.columns
```

```python
plt.figure(figsize = (12,6))

education_income.plot(kind = "bar" , stacked = True , xlabel = "Education" , ylabel = "Proportion" , title = "Relationship between education and income")
plt.legend(title = "Income" , loc = "upper right")
plt.show()
```

**or**

```python
b = px.bar(education_income , labels = {"education":"Education level" , "value":"Proportion" , "salary":"Income"} , title = "Relationship between education and income")
b.show()
```

# Q22 What is the distribution of occupation types among different income groups?

```python
df.columns
```

```python
occupation_income = pd.crosstab(df["occupation"] , df["salary"])
occupation_income
```

```python
proportion_occupation_income = occupation_income.div(occupation_income.sum(1), axis = 0)
proportion_occupation_income
```

```python
plt.figure(figsize = (12,8))
proportion_occupation_income.plot(kind = "bar" , stacked = True  )
plt.xlabel('Occupation')
plt.ylabel('Proportion')
plt.title('Distribution of Occupation Types Among Income Groups')
plt.legend(title='Income', bbox_to_anchor=(1, 1))
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

**or**

```python
boxplot = px.bar(proportion_occupation_income , title = "Distribution of Occupation Types Among Income Groups", labels = {"value":"Proportion","salary":"Income"} )
boxplot.show()
```
